```python
# get the cleaned data using pyspark
from pyspark.sql import SparkSession

'''
Initiate the spark session and read the csv file.
Then, split the data into fraud and non fraud category records.
'''
spark = SparkSession.builder.appName("FraudDetection").getOrCreate()
cleaned_data = spark.read.csv("../data/cleaned_credit_card.csv", header=True, inferSchema=True)
non_fraud_data = cleaned_data.filter(cleaned_data.Class == 0)
fraud_data = cleaned_data.filter(cleaned_data.Class == 1)
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
'''
We need to split the cleaned data into training and testing data.
We will use train_test_split function from sklearn.
Before that we need convert the pyspark dataframe input to pandas for the function to work.
'''
training_fraud, test_fraud = train_test_split(fraud_data.toPandas(), train_size = 0.7, shuffle=True)
training_non_fraud, test_non_fraud = train_test_split(non_fraud_data.toPandas(), train_size = 0.7, shuffle=True)
display(training_fraud) 
display(test_fraud) 
```
